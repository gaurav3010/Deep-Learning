# Autograd

```
x = torch.ones([3, 2], requires_grad=True)
print(x)
```

```
y = x + 5
print(y)
```

```
z = y*y + 1
print(z)
```

```
t = torch.sum(z)
print(t)
```

```
t.backward()
```

```
print(x.grad)
```

```
x = torch.ones([3, 2], requires_grad=True)
print(x)
y = x + 5
print(y)
r = 1/(1 + torch.exp(-y))
print(r)
s = torch.sum(r)
s.backward()
print(x.grad)
```

# Autograd example for W and B

```
x = torch.randn([20, 1], requires_grad=True)
y = 3*x - 2
```

```
w = torch.tensor([1.], requires_grad=True)
b = torch.tensor([1.], requires_grad=True)

y_hat = w*x + b

loss = torch.sum((y - y_hat) ** 2)
```

```
print(loss)
```

```
loss.backward()
```

```
print(w.grad, b.grad)
```

# In loop

```
learning_rate = 0.01

w = torch.tensor([1.], requires_grad=True)
b = torch.tensor([1.], requires_grad=True)

x = torch.randn([20, 1])
y = 3*x - 2

print(w.item(), b.item())

for i in range(10):
   
  y_hat = w*x + b
  loss = torch.sum((y_hat - y) ** 2)
  
  loss.backward()
  
  with torch.no_grad():
    w -= learning_rate * w.grad
    b -= learning_rate * b.grad
    
    w.grad.zero_()
    b.grad.zero_()
    
  print(w.item(), b.item())
```

# For large problem

```
%%time
learning_rate = 0.001
N = 10000
epochs = 2000

w = torch.rand([N], requires_grad=True)
b = torch.ones([1], requires_grad=True)

print(torch.mean(w).item(), b.item())
#print(y)

for i in range(epochs):
  
  x = torch.randn([N])
  y = torch.dot(3*torch.ones([N]), x) - 2

  y_hat = torch.dot(w, x) + b
  loss = torch.sum((y_hat - y) ** 2)
  
  loss.backward()
  
  with torch.no_grad():
    w -= learning_rate * w.grad
    b -= learning_rate * b.grad
    
    w.grad.zero_()
    b.grad.zero_()
      
  print(torch.mean(w).item(), b.item())
```
